```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3]
y=[2,4,1]
```

```python
plt.scatter(x,y)  # used to plot the points
plt.show()
```

```python
plt.plot(x,y)  # used to draw lines for the points
plt.show()
```

```python
x1=[1,2,3]
y1=[2,4,1]  #line1 points

x2=[1,2,3]
y2=[4,1,3]  #line2 points
```

```python
plt.plot(x1,y1,label="line 1")
plt.plot(x2,y2,label="line 2")
plt.legend() #distinguish lines
plt.show()
```

```python
plt.style.available
```

```python
plt.style.use('fivethirtyeight')  #grid like structure i.e,,  square boxes inside graph
x=[2,2,3,4,5,6]
y=[2,4,3,5,2,6]
plt.plot(x,y,color='blue',linestyle='dashed',linewidth=3,marker='o',markerfacecolor='green',markersize=8)
```

```python
# setting x and y axis range
plt.xlim(4,9)
plt.ylim(2,6)
```

```python
xleft=[1,2,3,4,5] # left side x-coordinates of the bar
height=[10,24,36,40,5]
t_label = ['one','two','three','four','five']
plt.bar(xleft,height,tick_label=t_label,width=0.5,color=['red','blue'])
```
